## Description
Independent Variable Name:

id: Unique identifier of each diamond

Carat: Carat weight of the cubic zirconia.unique weight of weight measurements used exckusively to weigh gemstones and diamonds

Cut : Quality of Diamond cut

Color : Colour of diamond

Clarity : Diamond quality is measure of purity and rarity of the stone ,graded by the visibility of these characheristic under 10 power magnification.

Depth: The depth of diamond is its heights (in millimeter) measured from culet(bottom tip) to the table (flat ,top surface)

Table :A diamonds table is the facet which can be seen when the stone is viewed faced up.

X: Diamond X dimention

Y :Diamond Y dimention

Z :Diamond Z dimention

## Target Variable :
Price:the Price of the cubic zirconia.
